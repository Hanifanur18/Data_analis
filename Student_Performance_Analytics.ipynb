# **Data Preparation**

```python
import pandas as pd
```

```python
file_url = 'https://raw.githubusercontent.com/Hanifanur18/Data_analis/main/StudentsPerformance%20.csv'
df = pd.read_csv(file_url)
```

```python
def set_average(math, read, write):
    return (math+read+write)/3
df['average score'] = set_average(df['math score'], df['reading score'], df['writing score'])
```

```python
print(df.info())
```

# **Data Visualization**

```python
d = df.pivot_table(
    index = 'parental level of education',
    columns = 'test preparation course',
    values= 'average score', aggfunc = 'mean'
)
```

```python
d.plot(kind = 'bar')
```

```python
a = df.groupby(by = 'parental level of education')['test preparation course'].value_counts()
print(a)
```
